```python
heights = [500]
run_time = -360
dates = pd.date_range("2010-02-01", freq="24H", end="2010-02-28")

stations = {
    "Davis": (-69, 78),
    "South Pole": (-90, 335),
    "Neumayer": (-71, 352),
    "Marambio": (-64, 303),
    "Maitri": (-70, 11),
}
```

```python
met = "ncep"
metdir = "/home/pankaj/Desktop/met/"
outbase = "/home/pankaj/Desktop/out/"
working = "/home/pankaj/phd/code/hytraj/working/"
cdir = "/home/pankaj/Desktop/control/"
hysplit = "/home/pankaj/phd/code/hytraj/working/hyts_std"

for height in heights:
    outdir = outbase + str(height) + "/"
    if not os.path.exists(outdir):
        os.mkdir(outdir)
    hy = HyControl(
        stations,
        height,
        run_time,
        working,
        metdir,
        outdir,
        met_type=met,
        exe=hysplit,
    )
    hy = hy.run(dates, vertical=0, model_top=10000.0, cdir=cdir)
```

```python
cfiles = glob.glob(cdir + "2010*")
hy = HyParallel(cfiles, ncpu=7)
hy.run()
```
